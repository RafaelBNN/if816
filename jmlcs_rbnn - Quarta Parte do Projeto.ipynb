# 1. Resumo

No final de 2019, o mundo foi surpreendido com o surgimento e a acelerada disseminação de um novo vírus: o SARS-CoV-2, causador da Covid-19, uma doença com primeiro caso detectado na China, mas que rapidamente se espalhou por todos os países do mundo. Desde então, toda a população mundial vem tentando buscar maneiras de parar - ou, pelo menos, desacelerar - a propagação desse vírus. Nesse contexto, a utilização de modelos matemáticos para prever o avanço da doença em diferentes cenários se mostra imprescindível, a fim de que sejam tomadas as medidas necessárias e mais efetivas no combate ao vírus. Assim, vamos propor uma variante do já conhecido modelo matemático SIR para prevenção de doenças, aplicado no estado de Pernambuco, para o Novo Coronavírus.

# 2. Introdução e Objetivos

O modelo matemático SIR, surgido no início do século XX, foi criado com o objetivo de estudar e prever o comportamento de doenças numa comunidade usando equações diferenciais. A ideia principal é dividir a população em 3 grupos: $S$, indicando a população suscetível a pegar a doença; $I$, indicando a população infecciosa; e $R$, que engloba os indivíduos que se recuperaram ou que morreram em decorrência da doença em questão.
<br><br>
Na situação atual da Covid-19, podemos tirar proveito desse modelo de modo que possamos prever o avanço da pandemia de acordo com algumas hipóteses iniciais. Nosso propósito será colher dados pertinentes sobre a pandemia no estado de Pernambuco para tentar predizer a disseminação do vírus de acordo com uma variante do modelo SIR, com previsão para 450 dias após o primeiro caso registrado no Estado.
<br><br>
A variante do modelo SIR empregada neste trabalho será a SEIR. A diferença da versão SEIR para a SIR é que o modelo SEIR básico adiciona um novo grupo: o grupo $E$ de indivíduos expostos. Esses indivíduos são caracterizados por terem contraído o vírus, mas ainda não terem demonstrado sintomas da doença. Essa mudança é importante pois indivíduos que já demonstraram sintomas têm uma probabilidade maior de contaminar outras pessoas.
<br><br>
Além de utilizar o modelo SEIR, buscaremos refinar o modelo tendo em mente o efeito de algumas medidas, como: medidas de higiene, quarentena de infectados e vacinação. Com o fim de esclarecer ainda mais os resultados obtidos, incluiremos, também, o grupo $O$ de indivíduos que vieram a óbito em consequência da doença. Por último, vamos analisar as conclusões a que nosso modelo chegou e comparar com o real cenário que vem ocorrendo no estado de Pernambuco.
<br><br>
Dessa forma, será necessário adaptar as equações do modelo SEIR inicial com o intuito de aprimorar o modelo de acordo com os as melhorias apresentadas.

# 3. Métodos

## Equações

### Modelo Inicial

Como dito anteriormente, o modelo SIR é baseado em 3 grupos. Supondo a população constante e chamando de $\beta$ a taxa de crescimento da doença e de $\gamma$ a taxa média de remoção, isso nos leva às equações:
- $dS/dt =−\beta*S*I/N$
- $dI/dt = \beta*S*I/N - \gamma*I$ 
- $dR/dt = \gamma*I$

É possível observar ver que a soma total $dS/dt + dI/dt + dR/dt$ dá zero, já que é feita a suposição de que a população não varia. Além disso, analisando termo por termo, notamos que a taxa de variação de indivíduos suscetíveis depende do contato que o grupo $S$ tem com o grupo $I$. Essa observação será relevante a diante.
<br><br>
Para o modelo SEIR, devemos adicionar uma nova equação para o novo grupo $E$. Agora, chamando de $\delta$ o tempo que um indivíduo infectado leva para se tornar infeccioso, temos:
- $dS/dt = -\beta*S*(I + q*E)/N$
- $dE/dt = \beta*S*(I + q*E)/N - E/\delta$
- $dI/dt = E/\delta - \gamma*I$
- $dR/dt = \gamma*I$

Aqui, observe o surgimento do grupo $E$ na equação do grupo $S$. Isso acontece porque um indivíduo exposto, apesar de não demonstrar sintomas, pode espalhar o vírus. A constante $q$ vem justamente para contabilizar a menor probabilidade que um indivíduo exposto tem de infectar outra pessoa, comparado à probabilidade de um indivíduo infeccioso de fazer o mesmo.
<br><br>
Para sofisticar o já conhecido modelo SEIR, será necessário acrescentar algumas novas variáveis. Vamos fazer isso nos subtópicos posteriores. 

### Óbitos

Inicialmente, vamos considerar o grupo de indivíduos que vieram a óbito em decorrência da doença. Note que, assim como o grupo $E$ era um subgrupo de $I$ no modelo SIR inicial, aqui, o grupo $O$ era um subgrupo de $R$. Desse modo, chamando de $\alpha$ a taxa de letalidade, teremos o seguinte modelo:

- $dS/dt = -\beta*S*(I + q*E)/N$
- $dE/dt = \beta*S*(I + q*E)/N - E/\delta$
- $dI/dt = E/\delta - \gamma*I - \alpha*I$
- $dR/dt = \gamma*I$
- $dO/dt = \alpha*I$

Dessa forma, observe que o grupo $R$ agora engloba apenas os indivíduos _recuperados_ (em vez de removidos), e $\gamma$ torna-se a taxa de _recuperação_ (em vez de remoção).
Note que a equação que define o novo grupo $O$ é $\alpha*I$. Isso se dá porque consideramos que todos os indivíduos expostos ($E$) eventualmente evoluirão para indivíduos infecciosos ($I$). Para continuarmos com a suposição de população constante, devemos subtrair esse termo $\alpha*I$ de alguma das outras equações. Como se trata de indivíduos infecciosos indo para o grupo de óbitos, subtraímos $\alpha*I$ da equação de $I$.

### Medidas de Prevenção

Agora, vamos considerar os aprimoramentos relacionadas a medidas de distanciamento social e higiene. Serão incluídos 2 novos parâmetros: $u$, que vai indicar a adoção de medidas protetivas pela população e $k$, que vai depender da proporção de indivíduos infecciosos que está "quarentenada". 
<br><br>
Incluiremos $u$ nas equações de modo que, à medida que $u$ aumenta (ou seja, a adoção de medidas protetivas aumenta), então a taxa de propagação do vírus ($\beta$) diminui. Podemos, então, considerar um termo $(1-u)$ multiplicando $\beta$ sempre que $\beta$ aparecer. Assim, $u$ assumirá valores no intervalo $[0,1]$ dependendo da adoção de medidas de distanciamento - com $u=1$ sendo o cenário ideal.
<br><br>
Para $k$, consideraremos que a taxa da população infecciosa que se isola tem impacto direto na taxa de recuperação $\gamma$. Assim, quanto maior $k$, maior $\gamma$. Similarmente a como fizemos com $u$, agora vamos adicionar um termo $(1+k)$ multiplicando $\gamma$. 
<br><br>
Assim:
- $dS/dt = -(1-u)*\beta*S*(I + q*E)/N$ 
- $dE/dt = (1-u)*\beta*S*(I + q*E)/N - E/\delta$
- $dI/dt = E/\delta - (1+k)*\gamma*I - \alpha*I$
- $dR/dt = (1+k)*\gamma*I$
- $dO/dt = \alpha*I$



### Vacinação

Hoje, a vacinação é a mais poderosa arma da humanidade contra a COVID-19. O motivo disso é que, enquanto medidas de distanciamento e quarentena de infectados se mostraram não muito sustentáveis, a vacinação não exige um grande esforço da população, contanto que aplicada de modo eficaz. Portanto, seria bastante interessante incluir essa questão no nosso modelo.
<br><br>
Quanto à forma com que a vacinação se aplica nas equações do modelo, a princípio pensamos em incluir essa medida apenas ao grupo dos suscetívis. Porém o que acontece, na verdade, é que são aptas a receber a vacina tanto pessoas que nunca tiveram contato com o vírus quanto pessoas que já foram, em algum momento, contaminadas. Desse modo, vamos adicionar mais um grupo (chamado de $V$, de vacinados) ao nosso modelo. Esse grupo será um grupo "à parte", pois podem existir indivíduos que pertençam, simultaneamente, ao grupo $R$ e ao grupo $V$, por exemplo.
<br><br>
As equações do modelo serão dadas por:

- $dS/dt = -(1-u)*\beta*S*(I + q*E)/N - \sigma*S$
- $dE/dt = (1-u)*\beta*S*(I + q*E)/N - E/\delta - \sigma*E$
- $dI/dt = E/\delta - (1+k)*\gamma*I - \alpha*I - \sigma*I$
- $dR/dt = (1+k)*\gamma*I$
- $dO/dt = \alpha*I$
- $dV/dt = \sigma*(S+E+I+R)$

Onde $\sigma$ indica a taxa de vacinação. Observe que subtraímos termos com $\sigma$ também das equações dos grupos $E$ e $I$, o que é discutível, dado que vacinas têm efeito diferente de soros: enquanto a imunização por vacina demora a ocorrer, a imunização por soro é quase instantânea.
<br><br>
Evidentemente, não podemos simplesmente introduzir essas equações da forma como estão nesse modelo. Isso implicaria dizer que a vacinação foi iniciada no mesmo dia em que a pandemia chegou em Pernambuco. Para considerar essa diferença das datas, incluiremos no modelo uma função degrau, que incorpora a vacinação nas equações apenas a partir de um certo ponto. Com isso, as equações a serem implementadas no modelo serão da forma:

- $dS/dt = -(1-u)*\beta*S*(I + q*E)/N - H(t-T)*\sigma*S$
- $dE/dt = (1-u)*\beta*S*(I + q*E)/N - E/\delta - H(t-T)*\sigma*E$
- $dI/dt = E/\delta - (1+k)*\gamma*I - \alpha*I - H(t-T)*\sigma*I$
- $dR/dt = (1+k)*\gamma*I$
- $dO/dt = \alpha*I$
- $dV/dt = H(t-T)*\sigma*(S+E+I+R)$

Onde $H$ é a função degrau e $T$ é o intervalo, em dias, desde a data em que o primeiro caso foi reportado até a data em que a vacinação começou.

Supondo a vacinação como linear, esse número de vacinados seria alcançado com uma taxa de vacinação $\sigma=0.824/500$. Isso nos dá $\sigma=0.0016$, que será o valor usado no nosso modelo.

# 4. Resultados e Discussão

Aplicando os dados obtidos nas equações descritas e utilizando códigos em Python, podemos obter previsões razoáveis da evolução da pandemia no estado de Pernambuco.
Utilizaremos, também, algumas bibliotecas de Python para visualização de gráficos e solução de EDOs, como MatPlotLib, SciPy e NumPy.

## Cenário 1

Inicialmente, vamos apenas aplicar os dados nas equações demonstradas. Vamos supor $u=0.15$ e $k=0.2$. Assim, teremos:

```python
# Importamos as bibliotecas necessarias
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt

# Implementacao da funcao degrau. Essa funcao retorna True apenas a partir do momento em que t ultrapassa t_inic (t_inic aqui seria o T em H(t-T)).
# Dessa forma, ao multiplicarmos um double por essa funcao, Python interpreta esse resultado como 0 (False) ou 1 (True).
def step(t, t_inic):
    return (t>t_inic)


# Introduzimos as informacoes obtidas
n_hab = 9616621         # Numero de habitantes de Pernambuco
beta = 0.58             # Taxa de infeccao para individuos infecciosos
q = 0.2                 # Taxa de infeccao para individuos expostos
delta = 5.1             # Tempo de incubacao do virus
gamma = 1/3.3           # Taxa de recuperação
alpha = 0.034           # Taxa de letalidade
sigma = 0.0016          # Taxa de vacinação

u = 0.15                # Descreve a adocao de medidas protetivas
k = 0.2                 # Descreve o efeito da quarentena de infectados

# Valores iniciais do modelo
e_inic = 0.0                                                    # Expostos
i_inic = 1.0                                                    # Infecciosos
r_inic = 0.0                                                    # Recuperados
o_inic = 0.0                                                    # Obitos
v_inic = 0.0                                                    # Vacinados
s_inic = n_hab-e_inic-i_inic-r_inic-o_inic-v_inic               # Suscetiveis

# Um array com as informacoes iniciais para ser usado na plotagem do grafico
x_inic = [s_inic,e_inic,i_inic,r_inic,o_inic,v_inic]

# A funcao a ser utilizada na exibicao do grafico
def projecao(x,t):   
    s, e, i, r, o, v = x
    dx = np.zeros(6)    # Inicializamos um array com 6 valores zero. Esse array vai conter as taxas de crescimento de cada grupo
    
    dx[0] = -(1-u)*beta*s*(i+(q*e))/n_hab - step(t,312)*sigma*s                     # dS/dt
    dx[1] = ((1-u)*beta*s*(i+(q*e))/n_hab) - (e/delta) - step(t,312)*sigma*e        # dE/dt
    dx[2] = (e/delta) - ((1+k)*gamma*i) - alpha*i - step(t,312)*sigma*i             # dI/dt
    dx[3] = (1+k)*gamma*i                                                           # dR/dt
    dx[4] = alpha*i                                                                 # dO/dt
    dx[5] = step(t,312)*sigma*(s+e+i+r)                                             # dV/dt

    return dx

# Inicializando os parametros necessarios para a exibicao
t = np.linspace(0,450,450)      # Como dito na Introducao, vamos calcular a previsao para 450 dias desde o registro do primeiro caso.
                                # Para visualizar o fim previsto da vacinacao, podemos aumentar esse numero para 850 e ver os 82.4% da 
                                # populacao no grupo dos vacinados.
x = odeint(projecao,x_inic,t)

s = x[:,0]
e = x[:,1]
i = x[:,2]
r = x[:,3]
o = x[:,4]
v = x[:,5]


# Funcoes para plotar o grafico
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, s/n_hab, 'green', lw=2, label='Suscetíveis')         # S
ax.plot(t, e/n_hab, 'orange',  lw=2,label='Expostos')           # E
ax.plot(t, i/n_hab, 'red', lw=2, label='Infecciosos')           # I
ax.plot(t, r/n_hab, 'magenta', lw=2, label='Recuperados')       # R
ax.plot(t, o/n_hab, 'black',  lw=2,label='Óbitos')              # O
ax.plot(t, v/n_hab, 'blue', lw=2, label='Vacinados')            # V


ax.set_xlabel('Dias')
ax.set_ylabel('Grupos (%)')
ax.set_ylim(-0.05,1.1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

Observamos, portanto, que nosso modelo de equações, junto aos dados obtidos, prevê o fim da pandemia em Pernambuco para aproximadamente 200 dias após o primeiro caso reportado em 12/03/2020. Isso quer dizer que a pandemia deveria ter acabado em setembro de 2020, com um total de infectados relativamente baixo. Esse cenário claramente não aconteceu. Um dos motivos dessa discrepância foi porque consideramos os parâmetros $u$ e $k$ com valores já dados inicialmente, o que diminuiu a taxa de contaminação $\beta$ e aumentou a taxa de recuperação $\gamma$ desde o início. Nos próximos cenários vamos tentar dar melhores aproximações.

```python
u = 0.0                 # Descreve a adocao de medidas protetivas
k = 0.0                 # Descreve o efeito da quarentena de infectados

# A funcao a ser utilizada na exibicao do grafico
def projecao(x,t):   
    s, e, i, r, o, v = x
    dx = np.zeros(6)    # Inicializamos um array com 6 valores zero. Esse array vai conter as taxas de crescimento de cada grupo
    
    dx[0] = -(1-u)*beta*s*(i+(q*e))/n_hab - step(t,312)*sigma*s                     # dS/dt
    dx[1] = ((1-u)*beta*s*(i+(q*e))/n_hab) - (e/delta) - step(t,312)*sigma*e        # dE/dt
    dx[2] = (e/delta) - ((1+k)*gamma*i) - alpha*i - step(t,312)*sigma*i             # dI/dt
    dx[3] = (1+k)*gamma*i                                                           # dR/dt
    dx[4] = alpha*i                                                                 # dO/dt
    dx[5] = step(t,312)*sigma*(s+e+i+r)                                             # dV/dt

    return dx

# Inicializando os parametros necessarios para a exibicao
t = np.linspace(0,450,450)      # Como dito na Introducao, vamos calcular a previsao para 450 dias desde o registro do primeiro caso.
                                # Para visualizar o fim previsto da vacinacao, podemos aumentar esse numero para 850 e ver os 82.4% da 
                                # populacao no grupo dos vacinados.
x = odeint(projecao,x_inic,t)

s = x[:,0]
e = x[:,1]
i = x[:,2]
r = x[:,3]
o = x[:,4]
v = x[:,5]


# Funcoes para plotar o grafico
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, s/n_hab, 'green', lw=2, label='Suscetíveis')         # S
ax.plot(t, e/n_hab, 'orange',  lw=2,label='Expostos')           # E
ax.plot(t, i/n_hab, 'red', lw=2, label='Infecciosos')           # I
ax.plot(t, r/n_hab, 'magenta', lw=2, label='Recuperados')       # R
ax.plot(t, o/n_hab, 'black',  lw=2,label='Óbitos')              # O
ax.plot(t, v/n_hab, 'blue', lw=2, label='Vacinados')            # V


ax.set_xlabel('Dias')
ax.set_ylabel('Grupos (%)')
ax.set_ylim(-0.05,1.1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

Aqui, observamos que houve um número consideravelmente maior no total de infectados e um maior número de mortos, passando de 10%. Isso mostra a importância das medidas de prevenção e quarentena.

## Cenário 3

Como mostrado anteriormente, nosso modelo se utiliza de uma função de degrau para a taxa de vacinação. Já que já tínhamos implementado a função, decidimos utilizá-la também sobre os parâmetros $u$ e $k$. O primeiro lockdown decretado em Pernambuco ocorreu de 16/05/2020 a 30/05/2020. Vamos considerar esses dias específicos como tendo medidas de prevenção tomadas mais corretamente ($u$ maior). Vamos supor $k=0.25$ e $u=0.3$ nesse momento. Isso vai nos dar:

```python
u = 0.3                # Descreve a adocao de medidas protetivas
k = 0.25                 # Descreve o efeito da quarentena de infectados

# A funcao a ser utilizada na exibicao do grafico. Aqui serao feitas as mudancas. Observe a diferenca sempre que u ou k aparecem
def projecao(x,t):   
    s, e, i, r, o, v = x
    dx = np.zeros(6)    # Inicializamos um array com 6 valores zero. Esse array vai conter as taxas de crescimento de cada grupo
    
    dx[0] = -(1-u*(step(t,65)-step(t,80)))*beta*s*(i+(q*e))/n_hab - step(t,312)*sigma*s                     # dS/dt
    dx[1] = ((1-u*(step(t,65)-step(t,80)))*beta*s*(i+(q*e))/n_hab) - (e/delta) - step(t,312)*sigma*e        # dE/dt
    dx[2] = (e/delta) - ((1+k*(step(t,65)-step(t,80)))*gamma*i) - alpha*i - step(t,312)*sigma*i             # dI/dt
    dx[3] = (1+k*(step(t,65)-step(t,80)))*gamma*i                                                           # dR/dt
    dx[4] = alpha*i                                                                                         # dO/dt
    dx[5] = step(t,312)*sigma*(s+e+i+r)                                                                     # dV/dt

    return dx

# Inicializando os parametros necessarios para a exibicao
t = np.linspace(0,450,450)      # Como dito na Introducao, vamos calcular a previsao para 450 dias desde o registro do primeiro caso.
                                # Para visualizar o fim previsto da vacinacao, podemos aumentar esse numero para 850 e ver os 82.4% da 
                                # populacao no grupo dos vacinados.
x = odeint(projecao,x_inic,t)

s = x[:,0]
e = x[:,1]
i = x[:,2]
r = x[:,3]
o = x[:,4]
v = x[:,5]


# Funcoes para plotar o grafico
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, s/n_hab, 'green', lw=2, label='Suscetíveis')         # S
ax.plot(t, e/n_hab, 'orange',  lw=2,label='Expostos')           # E
ax.plot(t, i/n_hab, 'red', lw=2, label='Infecciosos')           # I
ax.plot(t, r/n_hab, 'magenta', lw=2, label='Recuperados')       # R
ax.plot(t, o/n_hab, 'black',  lw=2,label='Óbitos')              # O
ax.plot(t, v/n_hab, 'blue', lw=2, label='Vacinados')            # V


ax.set_xlabel('Dias')
ax.set_ylabel('Grupos (%)')
ax.set_ylim(-0.05,1.1)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

Note que o "pico" do número de infectados ocorreu mais cedo que no cenário 1, como esperado, dado que as taxas de recuperação e contaminação não foram alteradas nos primeiros dias. É possível observar, também, que tivemos um maior número de óbitos neste cenário do que no cenário 1.

Comparando os resultados obtidos, principalmente nos cenários 1 e 3, é possível notar que o efeito de lockdowns impostos num curto período de tempo não têm um efeito tão claro, e é melhor tomar medidas de prevenção e isolamento mesmo que menos restritas, mas em longo prazo.

# 5. Conclusão

Infelizmente nosso modelo não teve ainda a complexidade suficiente para se aproximar do mundo real. Isso se dá, principalmente, pela tendência de modelos simples de considerar características bastante instáveis como constantes. Além da taxa de contaminação, outras taxas podem ser consideradas variáveis: a taxa de recuperação, dependendo da quantidade de pessoas "quarentenadas"; a taxa de letalidade, que pode diminuir com o tempo, com os profissionais de saúde desenvolvendo melhores tratamentos; a taxa de vacinação, que depende da disponibilidade de vacinas; entre outras.
<br>
Por isso, é de extrema importância que sejam realizados estudos mais aprofundados, que considerem tais variabilidades e até diferenças entre os grupos sociais.
<br>
Dito isso, podemos ver pelos gráficos reais que, embora a vacinação já ter iniciado há mais de 3 meses, os números de novos casos e óbitos não parecem diminuir. Isso mostra a importância das medidas de distanciamento e higiene, que, enquanto não tivermos 100% da população apta vacinada, devem ser respeitadas.
